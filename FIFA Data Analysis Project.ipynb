**Dropping Columns From Analysis**

```python
fifa.drop(['player_url','long_name','club_jersey_number','club_loaned_from','nation_jersey_number','player_face_url','club_flag_url','club_logo_url','nation_logo_url','nation_flag_url'],axis=1,inplace=True)
```

**Data Cleaning**

```python
#Numerical Features
fifa['international_reputation'].fillna(fifa['international_reputation'].mean(),inplace=True)
fifa['skill_moves'].fillna(fifa['skill_moves'].mean(),inplace=True)
fifa['weak_foot'].fillna(fifa['weak_foot'].mean(),inplace=True)
fifa['passing'].fillna(fifa['passing'].mean(),inplace=True)
```
